```python
import pubfig_loader
import matplotlib.pyplot as plt
import resface
import numpy as np
import umap
```

```python
pubfigfiles  =  pubfig_loader.load_files()
names = list(pubfigfiles.keys())
print(len(names))
name2idx = {}
for i in range(len(names)):
    name2idx[names[i]] = i
labels = []
files = []
for name in pubfigfiles.keys():
    name_idx = name2idx[name]
    for file in pubfigfiles[name]:
        labels.append(name_idx)
        files.append(file)

#the output of this block-- labels, files are the embedding of the person, and corrsponding file
```
