```python
import umap
import pickle
import matplotlib.pyplot as plt
```

```python
f = open('retina_arcface.pl','rb')
```

```python
embs_sim,embs_affine,embs_no,embs_rot,suc_labels  = pickle.load(f)
```

```python
def reduce_plot(emb):
    reducer = umap.UMAP(random_state=42)
    reducer.fit(emb)
    emb_reduced = reducer.transform(emb)
    plt.figure(dpi=500)
    plt.scatter(emb_reduced[:,0],emb_reduced[:,1],c=suc_labels,cmap='Spectral', s=1)
    plt.show()
```

```python
reduce_plot(embs_sim)
```

```python
reduce_plot(embs_affine)
```

```python
reduce_plot(embs_rot)
```

```python
reduce_plot(embs_no)
```
